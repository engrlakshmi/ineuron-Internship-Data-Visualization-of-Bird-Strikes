```python
num_people_injured=df.groupby('Number of people injured').size()
print(num_people_injured)
```

```python
def alt_bin(r):
    if r>=500:
        return 'High'
    elif r<=500:
        return 'Low'
      
```

```python
#Applying converted data into our dataset with new column - People_injured_Alitude

df['People_injured_Alitude']=df['Feet above ground'].apply(alt_bin)
df.head()
```

```python
Count_people=df.groupby('People_injured_Alitude')['Feet above ground'].count()
Count_people
```

```python
# No of pilots warned of bird strikes
num_pilots_warned=df.groupby('Pilot warned of birds or wildlife?').size()
print(num_pilots_warned)
```

```python
# Size of aircrafts affected by bird strikes
size_aircrafts=df.groupby('Is Aircraft Large?').size()
print(size_aircrafts)
```

```python
def aircraft_size(s):
    if s=='Yes':
        return 'Affected by bird strikes'
    elif s=='No':
        return 'Not affected by bird strikes'
```

```python
#Applying converted data into our dataset with new column - Aircraft_Size

df['Aircraft_Size']=df['Is Aircraft Large?'].apply(aircraft_size)
df.head()
```

```python
Count_aircrafts=df.groupby('Aircraft_Size')['Is Aircraft Large?'].count()
Count_aircrafts
```

# **Data Cleaning**

```python
# Handling missing data
df.isna()
```

```python
# To find null values
df.isnull().sum()
```

```python
# to find the duplicates
df.drop_duplicates(inplace=True)
duplicate=df[df.duplicated()]
print('There are '+ str(len(duplicate)) +'  duplicate values in the dataset.')
```

```python
# To find columns with missing values
columns_list=[]
for i in df.columns:
  if df[i].isnull().sum()>0:
    columns_list.append(i)
print('Columns having missing values are: ', columns_list)
```

```python
df['Aircraft: Airline/Operator'].fillna(method = 'ffill', inplace = True)
```

```python
mean = df['Feet above ground'].mean()
df['Feet above ground'].fillna(mean, inplace = True)
```

```python
df.isnull().sum()

```
